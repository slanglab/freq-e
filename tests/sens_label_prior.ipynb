counter-intuitive behavior when all the predictions are one-sided (>0.5, or <0.5) and the label prior is less than PCC 

```python
import numpy as np
import freq_e
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#set-up: all the test pred probs are <0.5, training prev also <0.5
#we get weird predictions when the training prev is less than 
#the mean of the test_pred_probs 
# I think this has to do with the ratios p(y=1|x)/pi and p(y=0|x)/(1-pi)
LABEL_PRIOR = 0.25
test_pred_probs = np.array([0.2, 0.3, 0.4])
log_odds = freq_e.estimate.calc_log_odds(test_pred_probs)
log_post_probs = freq_e.estimate.mll_curve(log_odds, LABEL_PRIOR)
print('LABEL_PRIOR=', LABEL_PRIOR)
print('PCC=', np.mean(test_pred_probs))
print('point=', freq_e.estimate.generative_get_map_est(log_post_probs))
plt.plot(freq_e.estimate.DEFAULT_THETA_GRID, log_post_probs)
plt.show()
```

```python
#if we set the label prior equal to PCC or greater than we're fine 
LABEL_PRIOR = 0.3
log_post_probs = freq_e.estimate.mll_curve(log_odds, LABEL_PRIOR)
print('LABEL_PRIOR=', LABEL_PRIOR)
print('point=', freq_e.estimate.generative_get_map_est(log_post_probs))
plt.plot(freq_e.estimate.DEFAULT_THETA_GRID, log_post_probs)
plt.show()
LABEL_PRIOR = 0.5
log_post_probs = freq_e.estimate.mll_curve(log_odds, LABEL_PRIOR)
print('LABEL_PRIOR=', LABEL_PRIOR)
print('point=', freq_e.estimate.generative_get_map_est(log_post_probs))
plt.plot(freq_e.estimate.DEFAULT_THETA_GRID, log_post_probs)
plt.show()
LABEL_PRIOR = 0.9
log_post_probs = freq_e.estimate.mll_curve(log_odds, LABEL_PRIOR)
print('LABEL_PRIOR=', LABEL_PRIOR)
print('point=', freq_e.estimate.generative_get_map_est(log_post_probs))
plt.plot(freq_e.estimate.DEFAULT_THETA_GRID, log_post_probs)
plt.show()
```

```python
#this same phenomena holds even with an extremely large number of docs 
LABEL_PRIOR = 0.25
test_pred_probs = np.random.uniform(low=0.1, high=0.5, size=10**6)
log_odds = freq_e.estimate.calc_log_odds(test_pred_probs)
log_post_probs = freq_e.estimate.mll_curve(log_odds, LABEL_PRIOR)
print('LABEL_PRIOR=', LABEL_PRIOR)
print('PCC=', np.mean(test_pred_probs))
print('point=', freq_e.estimate.generative_get_map_est(log_post_probs))
plt.plot(freq_e.estimate.DEFAULT_THETA_GRID, log_post_probs)
plt.show()
```
